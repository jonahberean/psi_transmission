## $^{\dagger\dagger}$ *(we connect both plexi guides together, as at the beginning)*

>[Sun, Dec 10, 2017, 22:30:50 ] prepare to break vacuum in order to perform more normalization measurements (we connect both plexi guides together, as at the beginning).

I'm wondering if these runs at the end are also invalid, for the same reason as the runs at the beginning are invalid. If the plexi-glass guides have been connected directly together, then the diameter difference problem is re-introduced. I'm referring to what Edgard describes in his correspondence to me:

> we used two different configurations in order to extract the transmission. Actually, only one is correct... You want to know how the presence of your test guide change your nb_of_counted_UCNs. That means the only difference between the two configurations (w and wo the guide you want to characterize) must be the presence of this guide, and nothing more. The first way to normalize was not correct because the effect of the adaptive flanges (from plexy to test guide, different diameters) was ignored. The only way to add the effect of the flanges is to have the flanges...

Perhaps we can see the difference directly if we look at the adjacent normalization runs from this period of time. Edgard claims that the difference can be seen clearly in the runs at the start, when switching configurations from run 21 (bad normalization) to 22 (good normalization).

```python
data = np.loadtxt("../data_main/12/8/T081217_0021.tof", usecols = (0,1));
plt.plot(data[:,0] * 0.1, data[:,1]);
plt.xlabel('Time [s]');
plt.ylabel('UCN Counts');
counts_21 = np.sum(data[:,1])
print('Total UCN counts: {}'.format(counts_21))
plt.show()

data = np.loadtxt("../data_main/12/8/T081217_0022.tof", usecols = (0,1));
plt.plot(data[:,0] * 0.1, data[:,1]);
plt.xlabel('Time [s]');
plt.ylabel('UCN Counts');
counts_22 = np.sum(data[:,1])
print('Total UCN counts (run22:run21): {}'.format(counts_22))
plt.show()
print('Ratio of UCN counts: {}'.format(counts_22 / counts_21))
```

Now let's compare the final direct shot measurement under the normalization configuration with the two flanges, that's run 249 on the 10th of December, with the first direct shot for the connected plexi-glass guides, run 273 later that day:

```python
data = np.loadtxt("../data_main/12/10/T101217_0249.tof", usecols = (0,1));
plt.plot(data[:,0] * 0.1, data[:,1]);
plt.xlabel('Time [s]');
plt.ylabel('UCN Counts');
counts_249 = np.sum(data[:,1])
print('Total UCN counts: {}'.format(counts_249))
plt.show()

data = np.loadtxt("../data_main/12/10/T101217_0273.tof", usecols = (0,1));
plt.plot(data[:,0] * 0.1, data[:,1]);
plt.xlabel('Time [s]');
plt.ylabel('UCN Counts');
counts_273 = np.sum(data[:,1])
print('Total UCN counts (run249:run273): {}'.format(counts_273))
plt.show()
print('Ratio of UCN counts: {}'.format(counts_249 / counts_273))
```

It's suggestive that in both cases the plexi-glass guide configuration has less counts. If the UCN source is degrading over time, then that effect would be doubled by this comparison. By "doubled" I mean that in one case I'm computing a later_run:earlier_run ratio, and in the latter case I'm computing an earlier_run:later_run ratio. Degradation of the source over time would could explain this difference in the ratios found here.
