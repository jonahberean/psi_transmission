# Sorting Data

I've sorted all of the existing main detector run data into a dedicated folder, **main_detector_sorted** for analysis. Part of this involved pruning the full data set for runs that are actually worth using in analysis. This pruning was carried out by cross-referencing the information in the elog, and some plotting of the data itself. 

The naming convention adopted for both the .tof and .txt files is **TDDMMYY_XXXX_SIZE_CONF_TYPE**, where **SIZE** is either '85mm' or '72mm', **CONF** labels the configuration(either 'norm' or a label referencing a specific guide), and **TYPE** labels the run type. The options are given below.

e.g.
**TDDMMYY_XXXX_85mm_norm_005s**

**SIZE_CONF**:

- 85mm_norm - the normalization, or guide-less configuration for 85 mm
- 72mm_norm - the normalization, or guide-less configuration for 72 mm
- 72mm_jpti - JP Ti guide with NiP
- 72mm_jpsu - JP SUS guide with NiP
- 85mm_ug01 - UGD01 guide (Electro-polished SUS guide)
- 85mm_ug03 - UGD03 guide (SUS hand-polished guide)
- 85mm_ug19 - Electro-polished SUS guide with NiP

**TYPE**:

- 005s - 5 second storage measurement
- 020s - 20 second storage measurement
- 100s - 100 second storage measurement
- shot - direct shot measurement
